<h1>### DATA ###</h1> <br> 
<h3>Retrieving data from <b><i>Freelancer.com API</i></b>
Then with the data we will construct a PANDAS DataFrame. <br>

    We will select those atributes of the data to work with. </h3>

```python
#Lets load the data from freelancing.com, Projects. 
import urllib, urllib2
import json
url = 'http://api.freelancer.com/Project/Search.json?nonpublic=0'
response = urllib2.urlopen(url)
data = json.load(response)

#data keeps a list of projects 
data = data['projects']['items']
```

```python
#Lets introduce the data from Freelancing.com into a DataFrame
from pandas import *
df = DataFrame(data)

#REVIEW: show dataframe properties, methods, and what efficient operation we can do to handle data.
```

```python
df['bid_count'] = df.bid_stats.apply(lambda x: x['count'])
```

```python
#REVIEW: show describe over workroom_comm, or isHourly
df.workroom_comm.describe()
```

```python
#Select only those columns
df_clean = df.filter(items=['short_descr','jobs','budget','budgetPeriod','bid_count'])

#CHECKOUT: FREELANCER.COM DATA OF API IS NOT IN FIRST NORMAL FORM. JobsDetails its a list of objects. 
corpus = df_clean.short_descr

#We add new columns to the DataFrame, to work on budget_max, budget_min
df_clean['budget_max'] = df_clean.budget.apply(lambda x: float(x['max']))
df_clean['budget_min'] = df_clean.budget.apply(lambda x: float(x['min']))

```

```python
"""
Show how to modify the coordinate formatter to report the image "z"
value of the nearest pixel given x and y
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

X = 10*np.random.rand(10,3)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(X, cmap=cm.jet, interpolation='nearest')

numrows, numcols = X.shape
def format_coord(x, y):
    col = int(x+0.5)
    row = int(y+0.5)
    if col>=0 and col<numcols and row>=0 and row<numrows:
        z = X[row,col]
        return 'x=%1.4f, y=%1.4f, z=%1.4f'%(x, y, z)
    else:
        return 'x=%1.4f, y=%1.4f'%(x, y)

ax.format_coord = format_coord
plt.show()
```

<h1>### Clusterization of projects using KMeans ###</h1>
<h3> we will cluster all the projects retrieve from <b><i> freelancer.com </i></b>. The idea is to generate labels for this clusters in which then we can bucket a new projects coming. In this way, a new project posted can be identified similar to all those into some cluster. </h3>

```python
from sklearn.cluster import KMeans
from numpy.random import RandomState
import numpy as np
rng = RandomState(1)

kmeans = KMeans(3, random_state=rng).fit(dataset)

np.round(kmeans.cluster_centers_, decimals=2)
print kmeans.labels_[:10]
print kmeans.labels_[20:30]
```

```python
# Lets traverse all the clusters and for each cluster show which projects are inside. 
from collections import defaultdict

clusters = defaultdict(list)

for index, item in enumerate(kmeans.labels_[:30]):
    clusters[item].append(df_clean.ix[index])

for key in clusters:
    print '-------------------'
    print key
    print clusters[key]
    
```
